```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
data=pd.read_csv("score.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

X=data[['Hours']]
y=data['Scores']
model=LinearRegression()
model.fit(X,y)
y_pred=model.predict(X)

print("MAE =",mean_absolute_error(y,y_pred))
print("MSE =",mean_squared_error(y,y_pred))
print("RMSE =",np.sqrt(mean_squared_error(y,y_pred)))
hrs=float(input("Enter your study hours ="))
final_result=round(model.predict([[hrs]])[0],1)
print(f"according to your {hrs} hours study your appriximate marks is {final_result}")

r2=r2_score(y,y_pred)
print("r2 score =",round(r2,1))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(11,7))
plt.scatter(data['Hours'],data['Scores'])
plt.plot(data['Hours'],y_pred,color='red',label='Marks based on hours')
plt.title("Students Marks Predictor")
plt.xlabel("Study Hours")
plt.ylabel("Obtained Marks")
plt.legend()
plt.grid()
plt.show()
```
