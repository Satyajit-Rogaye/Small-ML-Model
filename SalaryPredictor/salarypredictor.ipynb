SMALL LINEAR REGRESSION MODEL BASED ON PERSON JOB EXPERIENCE IT WILL PREDICT SALARY

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
data=pd.read_csv("studentdata.csv")
data.head()
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

model=LinearRegression()
X=data[['experience']]
y=data['salary']
model.fit(X,y)
y_pred=model.predict(X)

print("MAE =",mean_absolute_error(y,y_pred))
print("MSE =",mean_squared_error(y,y_pred))
print("RMSE =",np.sqrt(mean_squared_error(y,y_pred)))
expe=float(input("Enter your work experience ="))
final_result=round(model.predict([[expe]])[0],1)
print(f"according to your work experience {expe} years your appriximately salary is {final_result}")

r2=r2_score(y,y_pred)
print("r2 score =",round(r2,1))
```
